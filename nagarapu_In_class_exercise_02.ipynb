Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):

from bs4 import BeautifulSoup
import requests
import pandas as pd

data = {
    'rank': [],
    'name': [],
    'year': [],
    'runtime': [],
    'genre': [],
    'rating': [],
    'description': [],
    
}



```

```python


root_url = 'https://www.imdb.com/'


url = 'https://www.imdb.com//search/title/?groups=top_1000&sort=user_rating,desc&start=1'


def get_page_soup_data(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    return soup

def get_next_page_url(soup):
    pages_data = soup.find('div', class_='desc')
    if soup.find('div', class_='desc').find('a', class_='lister-page-next next-page'):
        nexturl = root_url + str(soup.find('div', class_='desc').find('a', class_='lister-page-next next-page')['href'])
        return nexturl
    else:
        return

while True:
   
    soup = get_page_soup_data(url)
    
   
    movielist = soup.find('div', class_='lister-list').find_all('div', class_='lister-item mode-advanced')
    
    
    for movie in movielist:
        name = movie.find('h3', class_ = 'lister-item-header').a.text.strip()
        rank = movie.find('h3', class_='lister-item-header').text.split('.')[0].strip()
        year = movie.find('h3', class_='lister-item-header').find('span', class_='lister-item-year text-muted unbold').text.strip('()')

        runtime = movie.find_all('p', class_='text-muted')[0].find('span', class_='runtime').text.strip()
        genre  = movie.find_all('p', class_='text-muted')[0].find('span', class_='genre').text.strip()
        rating = movie.find('div', class_='inline-block ratings-imdb-rating').text.strip()
        description = movie.find_all('p', class_='text-muted')[1].text.strip()

        data['rank'].append(rank)
        data['name'].append(name)
        data['year'].append(year)
        data['runtime'].append(runtime)
        data['genre'].append(genre)
        data['rating'].append(rating)
        data['description'].append(description)
        
    

    url = get_next_page_url(soup)

    print("Next URL: ", url)
    print("Data collected till previous url: ", len(data['rank']))

    
    if not url:
        break
    
    
```

```python
df = pd.DataFrame(data)
df.shape
```

The articles should be published in the last 10 years (2012-2022).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):




import pandas as pd
import requests
import json


topic = 'information retrieval'


frames = []

for offset in range(0, 1000, 100):
    url = f"https://api.semanticscholar.org/graph/v1/paper/search?query={topic}&offset={offset}&limit=100&fields=url,title,venue,year,authors,abstract"

    payload={}
    headers = {}
    
    
    response = requests.request("GET", url, headers=headers, data=payload)
    
    data = json.loads(response.text)['data']


    df = pd.DataFrame(data)
    
   
    frames.append(df) 


main_df = pd.concat(frames)

main_df.reset_index(drop=True)


```
